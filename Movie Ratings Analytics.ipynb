```python
# movies[movies.Genre == 'Drama'].BudgetMillions 
## this filters films by drama, and then get the budget of those Dramas
plt.hist(movies[movies.Genre == 'Action'].BudgetMillions, bins = 15)
plt.hist(movies[movies.Genre == 'Drama'].BudgetMillions, bins = 15)
plt.hist(movies[movies.Genre == 'Thriller'].BudgetMillions, bins = 15)
plt.show()
```

```python
# The above is not what we want completely because
# they are overlapping each other
# we want them to stack one on top the other
```

```python
[movies[movies.Genre == 'Action'].BudgetMillions,
movies[movies.Genre == 'Drama'].BudgetMillions]

#put the genres you want into one list, then pass it in to plt.hist
```

```python
plt.hist([movies[movies.Genre == 'Action'].BudgetMillions,movies[movies.Genre == 'Drama'].BudgetMillions], bins = 15)
plt.show()
```

```python
# we have the two genre side-by-side, but they are not stacked, which is what we want
```

```python
plt.hist([movies[movies.Genre == 'Action'].BudgetMillions, \
          movies[movies.Genre == 'Drama'].BudgetMillions], \
          bins = 15, stacked=True)
plt.show()
# use the stacked=True parameter to stack the histograms
```

```python
# we need to iterate the genres, find a way to populate the plt.hist with all genres using code
```

```python
list1 = list()
labels = list()
for gen in movies.Genre.cat.categories:  
    labels.append(gen)  #loops through genres, and adds it to the empty list 'labels'
    list1.append(movies[movies.Genre == gen].BudgetMillions) #loops through and adds the budget for each diff genre, 
                                                             #puts it in the empty list called 'list1'
plt.hist(list1, bins = 30, stacked=True, rwidth=1, label = labels) #parses both new lists into plt.hist
plt.legend() # uses 'label' parameter and 'labels' list to show the legend on the plot
plt.show() # shows histogram
```

```python
#Styling dashboards
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams['figure.figsize'] = 8,4
plt.rcParams["patch.force_edgecolor"] = True
import warnings
warnings.filterwarnings('ignore')
```

```python
f, axes = plt.subplots(2,2, figsize = (15,15))
sns.set_style("darkgrid", {'axes.facecolor':'black'})
#Plot 1
k1 = sns.kdeplot(movies.BudgetMillions, movies.AudienceRating, \
                 shade=True, shade_lowest=True, cmap = 'inferno', \
                 ax = axes[0,0])
k1b = sns.kdeplot(movies.BudgetMillions, movies.AudienceRating, \
                 cmap = 'cool', ax = axes[0,0])

#Plot 2
k2 = sns.kdeplot(movies.BudgetMillions, movies.CriticRating,   \
                 shade=True, shade_lowest=True, cmap = 'inferno', \
                 ax = axes[0,1])
k2b = sns.kdeplot(movies.BudgetMillions, movies.CriticRating,   \
                 cmap = 'cool', ax = axes[0,1])

# Plot 3
v1 = sns.violinplot(data=movies,       \
                    x = 'Year', y = 'BudgetMillions', ax = axes[1,0], \
                    palette = "YlOrRd")

#Plot 4
kde1 = sns.kdeplot(movies.CriticRating, movies.AudienceRating, \
                   shade=True, shade_lowest=True, cmap = 'Blues_r', \
                   ax = axes[1,1])
kde2 = sns.kdeplot(movies.CriticRating, movies.AudienceRating, \
                   cmap = 'gist_gray_r', ax = axes[1,1])

#Set
k1.set(xlim = (-20,160))
k2.set(xlim = (-20,160))

plt.show()
```
